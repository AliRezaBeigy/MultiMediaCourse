# Description

Show the 8x8 base image of the DFT transform

# Import required dependencies

- Import cv2 for read images
- Import numpy to work with arrays
- Import matplotlib to show image for each step

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Calculate W

W is symmetric matrix create which is made by multiplying the numbers 1 to N so we can use meshgrid

```python
N = 8

k, n = np.meshgrid(np.arange(N), np.arange(N))
omega = np.exp(-2j * np.pi / N)
W = np.power(omega, k * n) / np.sqrt(N)

# WH = W*T
WH = np.transpose(np.conj(W))
```

# Create Base Image

Following the formula create Base Image usign W matrix and show it

```python
A = np.empty((N, N, N, N))
for k1 in range(N):
    for k2 in range(N):
        A[k1][k2] = W[k1][:].reshape((N, 1)) * WH[:][k2]
        

fig, subplot = plt.subplots(N, N, figsize=(N * 1.5, N * 1.5))
plt.tight_layout()

for k1 in range(N):
    for k2 in range(N):
        subplot[k1][k2].axis('off')
        subplot[k1][k2].imshow(A[k1][k2].real, cmap="gray")
        subplot[k1][k2].set_title(f"({k1}, {k2})")

plt.show()
```
