## Question 1 Part 2

# Description

Apply fourier transform to the vertical and check board images then show the real, imaginary, absolute and angle

# Import required dependencies

- Import cv2 for read images
- Import numpy to work with arrays
- Import matplotlib to show image for each step

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Read Image

```python
vertical_image = cv2.imread("../Images/Vertical.png", cv2.IMREAD_GRAYSCALE)
checker_board_image = cv2.imread("../Images/CheckerBoard.png", 
                                 cv2.IMREAD_GRAYSCALE)

f, subplt = plt.subplots(1,2,figsize=(10,5))
subplt[0].imshow(vertical_image, cmap='gray')
subplt[0].set_title("Vertical Image")
subplt[1].imshow(checker_board_image, cmap='gray')
subplt[1].set_title("Checker Board Image")
```

# Image In Frequency Domain

```python
vertical_image_freq = np.fft.fft2(vertical_image)
checker_board_image_freq = np.fft.fft2(checker_board_image)

fig, subplot = plt.subplots(2, 5, figsize=(20, 10))

subplot[0][0].axis('off')
subplot[0][0].set_title("Vertical Image")
subplot[0][0].imshow(vertical_image, cmap="gray")
subplot[0][1].axis('off')
subplot[0][1].set_title("Real")
subplot[0][1].imshow(vertical_image_freq.real, cmap="gray")
subplot[0][2].axis('off')
subplot[0][2].set_title("Imaginary")
subplot[0][2].imshow(vertical_image_freq.imag, cmap="gray")
subplot[0][3].axis('off')
subplot[0][3].set_title("Absolute")
subplot[0][3].imshow(np.absolute(vertical_image_freq), cmap="gray")
subplot[0][4].axis('off')
subplot[0][4].set_title("Angle")
subplot[0][4].imshow(np.angle(vertical_image_freq), cmap="gray")

subplot[1][0].axis('off')
subplot[1][0].set_title("BoardChecker Image")
subplot[1][0].imshow(checker_board_image, cmap="gray")
subplot[1][1].axis('off')
subplot[1][1].set_title("Real")
subplot[1][1].imshow(checker_board_image_freq.real, cmap="gray")
subplot[1][2].axis('off')
subplot[1][2].set_title("Imaginary")
subplot[1][2].imshow(checker_board_image_freq.imag, cmap="gray")
subplot[1][3].axis('off')
subplot[1][3].set_title("Absolute")
subplot[1][3].imshow(np.absolute(checker_board_image_freq), cmap="gray")
subplot[1][4].axis('off')
subplot[1][4].set_title("Angle")
subplot[1][4].imshow(np.angle(checker_board_image_freq), cmap="gray")
plt.show()
```
