# Description

Create squre mask with the of 10x10 in the center of the zero matrix with size of 100x100, move the squre and apply transform, in the other hand make squre bigger in 5 level then apply transform

# Import required dependencies

- Import cv2 for read images
- Import math to get the maximum number we can use as infinite
- Import numpy to work with arrays
- Import random to generate random number
- Import matplotlib to show image for each step

```python
import cv2
import math
import numpy as np
from random import randint
import matplotlib.pyplot as plt
```

# Read Image

```python
lena_image = cv2.imread("../Images/lena512.bmp", cv2.IMREAD_GRAYSCALE)


plt.imshow(lena_image, cmap='gray')
```

# Move Squre

move a 10x10 squre in 100x100 zero matrix in random location

```python
fig, subplot = plt.subplots(5, 5, figsize=(20, 10))

squre = np.ones((10, 10))

for i in range(5):
    mask = np.zeros((100, 100))

    x = randint(0, mask.shape[0] - squre.shape[0])
    y = randint(0, mask.shape[1] - squre.shape[1])

    mask[x:x + squre.shape[0],
         y:y + squre.shape[1]] = squre

    subplot[0][i].axis(False)
    subplot[0][i].imshow(mask, cmap="gray")
    subplot[0][i].set_title("Mask")

    mask_freq = np.fft.fftshift(np.fft.fft2(mask))

    subplot[1][i].axis(False)
    subplot[1][i].imshow(mask_freq.real, cmap="gray")
    subplot[1][i].set_title("Real")

    subplot[2][i].axis(False)
    subplot[2][i].imshow(mask_freq.imag, cmap="gray")
    subplot[2][i].set_title("Imaginary")

    subplot[3][i].axis(False)
    subplot[3][i].imshow(np.absolute(mask_freq), cmap="gray")
    subplot[3][i].set_title("Absolute")

    subplot[4][i].axis(False)
    subplot[4][i].imshow(np.angle(mask_freq), cmap="gray")
    subplot[4][i].set_title("Angle")

plt.show()
```

As result, when we move the squre, the angle of the image in frequency domain changed and real and imaginary and absolute are fix
